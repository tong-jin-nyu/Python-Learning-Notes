## Commands

```python
# Import
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris, load_wine

# Load data
iris = load_iris()
wine = load_wine()
# Load iris into a dataframe and set the field names
df1 = pd.DataFrame(iris['data'], columns=iris['feature_names'])
df2 = pd.DataFrame(wine['data'], columns=wine['feature_names'])
```

### Structure

```python
# Generate fancier outputs
display(df1.head())
display(df2.head())
df1.describe()
```

```python
# Summary
df1.info()
df2.info()
```

```python
# Subset
display(df1[df1['sepal length (cm)'] >= 7.0])
```

```python
# Join data
df_to_merge = df1.drop('sepal length (cm)', axis=1)
df_to_get_merged = df1[['petal length (cm)', 'petal width (cm)']]
df_merge = pd.merge(df_to_merge, df_to_get_merged, on='petal length (cm)')
display(df_merge.head())
```

### Data Visualization

```python
%matplotlib inline
plt.style.use('seaborn')

# Histogram
_ = plt.hist(df1['sepal length (cm)'])
_ = plt.xlabel('Sepal Length (cm)')
_ = plt.ylabel('Counts')
_ = plt.title('Histogram of Sepal Length (cm)')
plt.show()
```

```python
# Linear regression model fit
sns.regplot(x='sepal length (cm)', y='petal length (cm)', data=df1)
```
